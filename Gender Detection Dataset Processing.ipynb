```python
for i in range(len(img_id)):
    
    if (gen[i] == -1):
        shutil.copyfile('img_align_celeba/'+img_id[i],'Female/'+img_id[i])
    else:
        shutil.copyfile('img_align_celeba/'+img_id[i],'Male/'+img_id[i])

    if (i % 1000 == 0):
        print(i,'Iterations Done!')
```

```python
for i in range(len(img_id)):
    
    if (i < 160000):
        if (gen[i] == -1):
            shutil.copyfile('img_align_celeba/'+img_id[i],'Dataset/Train/Female/'+img_id[i])
        else:
            shutil.copyfile('img_align_celeba/'+img_id[i],'Dataset/Train/Male/'+img_id[i])

    elif (i >= 160000 and i < 180000):
        if (gen[i] == -1):
            shutil.copyfile('img_align_celeba/'+img_id[i],'Dataset/Test/Female/'+img_id[i])
        else:
            shutil.copyfile('img_align_celeba/'+img_id[i],'Dataset/Test/Male/'+img_id[i])
            
    else:
        if (gen[i] == -1):
            shutil.copyfile('img_align_celeba/'+img_id[i],'Dataset/Validation/Female/'+img_id[i])
        else:
            shutil.copyfile('img_align_celeba/'+img_id[i],'Dataset/Validation/Male/'+img_id[i])
        
    if (i % 1000 == 0):
        print(i,'Iterations Done!')
```
